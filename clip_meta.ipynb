```python
import os

processed_files = os.listdir('/home/chengxin/chengxin/Dataset_feature/chem_grid_lrs2_clip_b16')
processed_files = [os.path.splitext(file)[0] for file in processed_files]
len(processed_files)
```

```python
import csv
from tqdm import tqdm

def write_csv_line_by_line(filename, data):
        writer = csv.writer(csvfile)
        for row in data:
            writer.writerow(row)

# Example usage
data = chem_files + grid_files + lrs_files
c = 0

filename = '/home/chengxin/chengxin/Dataset_Sound/MetaData/vaflow2_meta/clip/visualtts_append.csv'
with open(filename, 'w', newline='', encoding='utf-8') as csvfile:
    writer = csv.writer(csvfile, delimiter=' ')
    for row in tqdm(data):
        basename  = row.split('/')[-1]
        basename = os.path.splitext(basename)[0]
        if basename in processed_files:
            continue
        c += 1
        adjusted_row = [row, -1, -1, -1, -1]
        writer.writerow(adjusted_row)
        

c
```
